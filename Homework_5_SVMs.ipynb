#### Questão 3
###### *Um classificador usa vetores de entrada com dimensão K=5 “features”. Após treinar uma SVM linear, o número de vetores de suporte foi de 450 exemplos. Neste caso, calcular o kernel linear corresponde a um produto interno entre dois vetores de dimensão K=5, cada, o que requer K multiplicações e K-1 adições. Estime o fator $F = Coriginal / Cperceptron$ de redução do custo computacional da etapa de teste ao converter esta SVM linear para um perceptron. Assuma que os custos $Coriginal$ e $Cperceptron$ correspondem ao número de multiplicações e adições usando-se, respectivamente, a SVM original com os 450 vetores de suporte e após sua conversão para perceptron.*
###### A função de decisão de uma SVM linear pode ser calculada da seguinte maneira:
$$f(x) = \sum_{i=1}^{N_s} \alpha_i y_i K(x_i, x) + b$$
###### Onde:
* ###### $Ns = 450$, sendo o número de vetores de suporte;
* ###### $K(x_i, x)$, é o produto interno dos vetores sendo o **kernel linear**.
###### De acordo com o comando, o produto interno entre dois vetores de dimensão K=5, requer **5 multiplicações** e $ K-1 $ adições = **4 adições**. Para cada um dos ${N_s}$  vetores de suporte, realizamos um produto interno (requer K mulitplicações) e o resultado é multiplicado pelo escalar $\alpha_i y_i$ (requer 1 multiplicação).
1. Custo Computacional da SVM Linear Original $({C_{original}})$ 
>> ###### **Multiplicações para a SVM Original**
* ###### Multiplicações para os produtos internos: ${N_s} . K = 450 . 5 = 2250$ 
* ###### Multiplicações para ponderar pelos ${N_s}$ termos $\alpha_i y_i$: ${N_s} . 1 = 450 . 1 = 450.$
* ###### Total multiplicações $ ({M_{original}}): 2250 + 450 = 2700.$
>> ###### **Adições para a SVM Original**
* ###### Adições para os produtos internos: ${N_s} . (K -1)  = 450 . 4 = 1800$ 
* ###### Adições para somar os ${N_s}$ termos: ${N_s} - 1 = 450 - 1 = 449.$
* ###### Adições para somar o termo de polarização: 1
* ###### Total adições $ ({A_{original}}): 1800 + 449 + 1 = 2250.$
**Custo total da SVM Original $${C_{original}} = 4950.$$** 


2. Custo Computacional do Perceptron (SVM Linear Convertida - $({C_{perceptron}})$)
###### Para uma SVM linear, o vetor de pesos $w$ pode ser calculado a partir dos vetores de suporte e os seus multiplicadores de Lagrange:
$$w = \sum_{i=1}^{N_s} \alpha_i y_i x_i$$
###### Calculando o vetor $w$, a classificação de um novo vetor $x$ se torna um produto interno, como em um perceptron:
$$f(x) = w . x + b$$
###### O vetor $w$ terá a mesma dimensão dos vetores de entrada, ou seja, $K=5$.
>> ###### **Multiplicações para o Perceptron**
* ###### Total multiplicações $({M_{perceptron}}): K = 5$
>> ###### **Adições para o Perceptron**
* ###### Adições para o perceptron será de $K-1 = 4$ mais o termo de polarização $b = 1$
* ###### Total adições $({M_{perceptron}}): K = 5.$
##### **Custo total do Perceptron $${C_{perceptron}} = 10.$$** 
3. Fator de Redução $(F)$
###### O fator de redução é definido como a razão entre o custo da SVM original e o custo do perceptron: 
$$F = \frac {Coriginal}{Cperceptron}$$ 
$$F = \frac {4950}{10} = 495$$
